```python
llm=ChatGroq(groq_api_key=api_key,model="llama-3.3-70b-versatile")
llm
```

```python
from langchain_core.messages import (
    AIMessage,
    HumanMessage,
    SystemMessage
)
```

```python
speech="Good evening everyone. Today, I stand before you to remind us of the power of unity and perseverance. In a world filled with challenges, it is our collective strength that allows us to rise above adversity. Each of us carries unique talents, and when we bring them together, we create something extraordinary. Let us embrace collaboration, celebrate diversity, and commit ourselves to building a brighter future. Success is not measured by individual triumphs alone, but by the progress we achieve together. With courage, compassion, and determination, there is no limit to what we can accomplish. Thank you.”"
```

```python
chat_message = [
    SystemMessage(content="You are an expert with expertise in summarizing speeches."),
    HumanMessage(content=f"Please provide a short and concise summary of the following speech:\nText: {speech}")
]
```

```python
token=llm.get_num_tokens(speech)
token
```

```python
llm.invoke(chat_message) #getting meta data as well

```

```python
llm.invoke(chat_message).content #getting only msg
```

##Prompt Tempkate Text Summarization 

```python
from langchain_core.prompts import PromptTemplate
from langchain_openai import ChatOpenAI
```

```python
generictemplate="""
write a summary of following speech:
speech:{speech}
Translate the precise summary to {language}
"""
```

```python
prompt=PromptTemplate(
    input_variables=['speech','language'],
    template=generictemplate
    )
prompt
```

```python
result=prompt.format(speech=speech,language='hindi')
result
```

```python
llm.get_num_tokens(result)
```

```python
chain2=prompt|llm
```

```python
summ=chain2.invoke({"speech":speech,"language":"hindi"})
print(summ)
```

```python
print(summ.content)
```
